## Matplotlib

Matplotlib is the most popular Python plotting package. It can create:

* charts and graphs
* animations
* interactive visualizations

It has two main modes:

* scripting mode: like giving commands to a Python turtle
* object-oriented mode: builds figures using objects

Scripting mode is easier to start with; object-oriented mode is more powerful and flexible.

## Matplotlib Vocabulary 

Matplotlib has a lot of its own vocabulary.

This [page](https://matplotlib.org/stable/tutorials/introductory/usage.html) describes commonly used terms. At the most basic level, we have:

* the canvas: the object that actually draws everything on the screen (you won't usually be directly working with this)
* the figure: your entire visualization or diagram, including the axes, its title, etc.
* the axes: your plotted data; there can be more than one set of axes in a figure (e.g., a figure with multiple line plots on it)
* each axis: with tick marks, etc.

## Basic Boilerplate Code

Using scripting mode (pyplot), all you have to do is:

1. import matplotlib
2. plot something
3. then show it.

```
import matplotlib.pyplot as plt
# your code
plt.show
```

##Line Graphs

Used for numeric or continuous data

In matplotlib:
```
plt.plot(x_values, y_values, format)
```

format is optional, but can specify line color, solid or dashed line, etc.



The following example plots a parabola: x and x squared.

Note that the axis and ticks are created automatically.

````python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(100)
y=x**2
plt.title("A Parabola")
plt.plot(x,y)
plt.show()
````

The following plots a couple of line graphs relating pizza size and number of toppings.

The xticks and the yticks take the locations of the ticks and the labels of the ticks.

xlabel and ylabel take the label to be put on each axis.

xlim and ylim specify the limits of each axis.

title gives the overall title of the graph.

````python
# another line graph example

x=[0,1,2,3,4] # pizza size
y=[0,2,4,9,16] # number of toppings
# store 2
x2=[0,1,2,3,4] # pizza size
y2=[4,5,6,7,8] # number of toppings
````

````python
# plot x against y with dashed red line
# plot x2 against y2 with solid blue line
plt.plot(x, y, 'r--', x2, y2, 'b-')

# set up the ticks and tick labels
plt.xticks([0, 2, 4], ['small (0)', 'medium (2)', 'large (4)'])
plt.yticks([0, 8, 16], ['not much (0)', 'some (8)','a lot (16)'])

# axis labels
plt.xlabel('pizza size')
plt.ylabel('number of toppings')
plt.title('Pizza Size v. Number of Toppings')


# axis limits
plt.xlim(-1, 5)
plt.ylim(0, 20)
plt.show()
````

## Legends

The following example shows a legend on a line graph. 

Here we hypothesize that the price of a pizza is a square function of the number of toppings.

We add a label to the plot commands that can be used in the legend.

````python
plt.plot(np.arange(10), color="blue", label="number of toppings")
plt.plot(np.arange(10), np.arange(10) ** 2, color="orange", label="price of pie")
plt.legend(loc="upper left")
plt.show()
````

## Scatter Plots

* plot individual (x,y) pairs
* sometimes a line graph is added to show a trend

````python
x = np.random.randint(1, 20, (10, ))
y = np.random.randint(10, 35, (10, ))
plt.scatter(x, y)
plt.show()
````

Annotating scatter plots:

````python
# to add labels, use annotate
labels = (chr(i) for i in range(65, 75))
plt.scatter(x, y)
# slight offset so that the labels aren't right on the point
# zip generates a tuple iterator
for x, y, label in zip(x, y, labels):
    plt.annotate(label, xy=(x+0.3, y-0.2))
plt.show()    
````

## Making Bar Graphs

Need categories which are the names of the bars

Counts associated with each bar

````python
# Setting up a bar graph
# bar categories
attitudes = ("Great", "OK", "Poor")

# counts for each bar
num_votes = [5, 3, 12]
# could also store the categories and counts in a dictionary
# the x values (based on number of labels)
x = np.arange(len(attitudes))
print(x)
````

````python
# create the bar chart; can use any RGB color in Hex
plt.bar(x, num_votes, align='center', color='#aaddff')
# set up the ticks for the categories
plt.xticks(x, attitudes)
# set up the y axis limits
plt.ylim(0, 14)
# x and y axis labels
plt.ylabel('Votes')
plt.xlabel('Attitude')
# overall title
plt.title('What do you think of this graph?')
plt.show()
````

Matplotlib supports various styles associated with other projects. These include:

* ggplot
* fivethirtyeight
* seaborn-dark
* seaborn-notebook
* bmh

To use a style, call "style.use" before using "plot" and "show".

````python
x=np.arange(0,25)
y=(x*4)**2
x2 = np.arange(0, 25)
y2 = x2 ** 3

# default style, don't do anythin
plt.plot(x, y, 'r-', x2, y2, 'b:')
plt.show()
````

````python
# bmh style
plt.style.use('bmh')
plt.plot(x, y, 'r-', x2, y2, 'b:')
plt.show()
````

## Pie Charts

```
plt.pie(sizes, explode=None, labels=None, autopct=None)
```
* sizes: the sizes of the slices
* explode: offset for the slice
* labels: labels for each slice
* autopct: format of value in slice



````python
edible_pies = ['strawberry', 'chocolate', 'apple', 'blueberry']
sizes = [10.2, 7.3, 4, 3]
plt.pie(sizes, labels=edible_pies)
#plt.pie(sizes)
plt.axis('equal') # deskews view
plt.show()
````

````python
plt.pie(sizes, labels=edible_pies, autopct=lambda x:f'{round(x,2)}%')
plt.axis('equal')
plt.show()
````

## Subplots

plt.subplot(rows, cols, index) 

Allow you to make multiple plots on a grid.

rows and cols are the dimensions of the grid.

index starts at 1 in the upper left corner and increases to the right.

So for a 2 by 2 grid, the index ranges from 1 to 4.

````python
plt.subplot(2, 2, 1)
# k stands for black
plt.plot(np.arange(10), 'k:')

plt.subplot(2, 2, 2)
plt.pie((1, 2, 3))
plt.axis('equal')

plt.subplot(2, 2, 3)
plt.bar(np.arange(4), (5, 7, 1, 2))

plt.subplot(2, 2, 4)
plt.plot(np.arange(10) ** 2, color="#ffaa00")
plt.show()
````

## Decrypting an encrypted file by examining the frequency of letters in it.

We will assume simple substitution of one letter for another.

````python
# let's look at the first 5 lines of the encrypted file
with open('/content/drive/MyDrive/0479 Spring 2022/data/enc.txt', 'r') as f:
    for i in range(5):
        print(f.readline(), end="")
````

````python
# let's see letter frequency
from collections import Counter
counter = Counter()
with open('/content/drive/MyDrive/0479 Spring 2022/data/enc.txt', 'r') as f:
    for line in f:
        for ch in line:
            letter = ch.lower()
            if letter.isalpha():
                counter[letter] += 1
                
most_common = counter.most_common()
freq_source = ''.join(t[0] for t in counter.most_common())

# plot in original order
plt.subplot(2, 1, 1)
plt.xticks(np.arange(len(counter)), counter.keys())
plt.bar(np.arange(len(counter)), counter.values(), 1)

# plot in frequency order
plt.subplot(2, 1, 2)
plt.xticks(np.arange(len(most_common)), [t[0] for t in most_common])
plt.bar(np.arange(len(most_common)), [t[1] for t in most_common], 1)
````

````python
# now substitute with letter frequency commonly found in texts
# found online, e most frequent, t, next, etc.
freq_target = 'etaonisrhldcmufpgwybvkjxqz'
# this gets us somewhat close, right!
with open('/content/drive/MyDrive/0479 Spring 2022/data/enc.txt', 'r') as f:
    l=0
    for line in f:
        s = ''
        for ch in line:
            letter = ch.lower()
            if letter.isalpha():
                i = freq_source.index(letter)
                new_letter = freq_target[i]
                s += new_letter
            else:
                s += ch
        print(s, end='')
        l+=1
        if l==10: break
````
